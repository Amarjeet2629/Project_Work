```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.DataFrame(pd.read_csv('/home/amarjeet-pc/Documents/Project_Work/Data.csv'))
df1 = pd.DataFrame()
df1 = df[1:]
df1 = df1.drop(columns = ['Unnamed: 0'])
df1.shape
```

```python
df1 = df1.set_index('Date')
colmn = df1.columns
#print (df1)
#print(colmn)
for i in colmn[1:]:
    #x = (min(df1[i]))
    (df1[i]/100).plot(figsize=(10,6))
#plt.colorbar()
plt.legend()
plt.xlabel('Date')
plt.show()
```

```python
avg_annual_return = ((df1.iloc[-1]/df1.iloc[0])**(1/567)-1)*(250/12)
avg_annual_return
```

# Portfolio_return

```python
weights =np.ones(18)
weights = weights/18
portfolio_return = np.dot(weights,avg_annual_return)
portfolio_return
```

```python
monthly_return = df1/df1.shift(1)-1
monthly_return.head()
covar_df1 = monthly_return.cov()*12
#covar_df1.head()
```

# Portfolio_risk

```python
df1_var = np.dot(weights.T, np.dot(covar_df1, weights))
df1_portfolio_risk = df1_var**0.5
df1_portfolio_risk
```

# Sharpe_ratio

```python
rf  = 0.02
df1_sharpe_ratio = ((portfolio_return)-rf)/df1_portfolio_risk
df1_sharpe_ratio
```
